```python
import pandas as pd

# Create a Series
marks = pd.Series([78, 85, 90, 66, 88])

print("Student Marks:")
print(marks)

# Basic operations
print("\nAverage Marks:", marks.mean())
print("Maximum Marks:", marks.max())
print("Minimum Marks:", marks.min())
```
